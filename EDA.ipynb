```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

amazon = pd.read_csv('AMAZON_weekly.csv')
amazon['Date']= pd.to_datetime(amazon['Date'])
amazon.describe()
```

```python
apple = pd.read_csv('APPLE_weekly.csv')
apple['Date']= pd.to_datetime(apple['Date'])
apple.describe()
```

```python
google = pd.read_csv('GOOGLE_weekly.csv')
google['Date']= pd.to_datetime(google['Date'])
google.describe()
```

```python
meta = pd.read_csv('META_weekly.csv')
meta['Date']= pd.to_datetime(meta['Date'])
meta.describe()
```

```python
netflix = pd.read_csv('NETFLIX_weekly.csv')
netflix['Date']= pd.to_datetime(netflix['Date'])
netflix.describe()
```

## Analysis of Stats Above
It appears that Meta has the highest closing price for stocks, followed by Netflix. However, Meta is the newest company, being founded in 2012. Its average is likely higher because of inflation. The other companies earn similarly, yet were founded earlier so their stocks initially cost less. 

Looking at the maximum closing price, this analysis does not seem to be accurate. Netflix has the highest maximum closing price of $1156.49, followed by Meta with $736.02. Google, Amazon, and Apple are all in the $200s for closing price, thus it does seem like in general, Meta and Netflix stocks are worth more. 

## Pretty Pictures

```python
companies = [apple, amazon, google, meta, netflix]
names = ['Apple', 'Amazon', 'Google', 'Meta', 'Netflix']

plt.figure(figsize=(10,7))
for i, company in enumerate(companies, 1):
    plt.subplot(2,3,i)
    plt.scatter(company['Date'], company['Close'], s=10)
    plt.title(names[i-1])
plt.tight_layout()
plt.suptitle('Closing Price since Founding', fontsize=14)
plt.subplots_adjust(top=0.9)
plt.show()

plt.figure(figsize=(10,7))
for i, company in enumerate(companies, 1):
    plt.subplot(2,3,i)
    plt.scatter(company['Date'][-104:], company['Close'][-104:], s=10)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b' '%Y'))
    plt.xticks(rotation=45)
    plt.title(names[i-1])
plt.tight_layout()
plt.suptitle('Closing Price over the last two years', fontsize=14)
plt.subplots_adjust(top=0.9)
plt.show()


plt.figure(figsize=(10,7))
for i, company in enumerate(companies, 1):
    plt.subplot(2,3,i)
    plt.scatter(company['Date'][-17:], company['Close'][-17:])
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b' '%Y'))
    plt.xticks(rotation=45)
    plt.title(names[i-1])
plt.tight_layout()
plt.suptitle('Closing Price since Trump took office', fontsize=14)
plt.subplots_adjust(top=0.9)
plt.show()


```

Although there has been a general increase in stock prices, espcially over the past couple of years (although maybe inflation has something to do with it), stock prices have gone down dramatically since Trump's Inauguration in January for all 5 companies

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = apple.corr()

plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap between Apple Features')
plt.show()
```

```python
correlation_matrix = amazon.corr()

plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap between Amazon Features')
plt.show()

correlation_matrix = google.corr()

plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap between Google Features')
plt.show()

correlation_matrix = meta.corr()

plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap between Meta Features')
plt.show()

correlation_matrix = netflix.corr()

plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap between Netflix Features')
plt.show()
```

```python
amazon['company'] = 'amazon'
google['company'] = 'google'
apple['company'] = 'apple'
netflix['company'] = 'netflix'
meta['company'] = 'meta'

result = pd.concat([amazon, google, apple, netflix, meta], axis=0)
result.tail()

result = pd.get_dummies(result, columns=['company', ])

result.replace({False: 0, True: 1}, inplace=True)

correlation_matrix2 = result.corr()

plt.figure(figsize=(9, 7))
sns.heatmap(correlation_matrix2, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap between Companies using One Hot Encoding')
plt.show()
```

```python
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler

names = ['Apple', 'Amazon', 'Google', 'Meta', 'Netflix']
scaler_ap = StandardScaler()
apple_norm = scaler_ap.fit_transform(apple[['Close']][-104:])
scaler_am = StandardScaler()
amazon_norm = scaler_am.fit_transform(amazon[['Close']][-104:])
scaler_ne = StandardScaler()
netflix_norm = scaler_ne.fit_transform(netflix[['Close']][-104:])
scaler_me = StandardScaler()
meta_norm = scaler_me.fit_transform(meta[['Close']][-104:])
scaler_go = StandardScaler()
google_norm = scaler_go.fit_transform(google[['Close']][-104:])

companies = [apple_norm, amazon_norm, google_norm, meta_norm, netflix_norm]


plt.figure(figsize=(10,7))
for i, company in enumerate(companies, 1):
    plt.subplot(2,3,i)
    plt.grid(True)
    plt.boxplot(company)
    plt.xlabel(names[i-1], fontsize=14)
    plt.title(names[i-1])
plt.tight_layout()
plt.suptitle('Standardized Spread', fontsize=14)
plt.subplots_adjust(top=0.9)
plt.show()
```
